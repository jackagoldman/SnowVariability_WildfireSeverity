Directed Acyclic Graphs for causal relationships

```python
import networkx as nx
from matplotlib import pyplot as plt
import graphviz # pygraphviz should be available
import pydot
```

```python
DG = nx.DiGraph()
DG.add_node("Biomass")
DG.add_node("Canopy Closure")
DG.add_node("Stand Age")
DG.add_node("Snow-free date")
DG.add_node("Snow-free duration")
DG.add_node("Burn Severity")
DG.add_node("Drought")
DG.add_node("Topography")

DG.add_edges_from([("Biomass", "Snow-free date"), ("Canopy Closure", "Snow-free date"), ("Stand Age", "Snow-free date"), ("Biomass", "Burn Severity"),
                    ("Canopy Closure", "Burn Severity"), ("Stand Age", "Burn Severity"), ("Snow-free date", "Snow-free duration"),
                    ("Topography", "Snow-free date"), ("Topography", "Burn Severity"), ("Drought", "Snow-free duration"), 
                    ("Drought", "Burn Severity"), ("Snow-free date", "Burn Severity"), ("Snow-free duration", "Burn Severity")])

nx.is_directed_acyclic_graph(DG)
```

```python
plt.tight_layout()
nx.draw_networkx(DG, arrows=True)

```

```python
#center_node = "Snow-free date"  # Or any other node to be in the center
#edge_nodes = set(DG) - {center_node}
pos = {"Biomass": (0,0),
      'Canopy Closure': (1,0),
      "Stand Age": (2,0),
      "Snow-free date": (1, -1),
       "Snow-free duration": (1,-2),
       "Burn Severity" : (1,-3),
       "Topography": (1.9,-1),
       "Drought": (0.1,-1),
      }



nx.draw(DG, font_weight='bold',  with_labels = True, pos = pos,  arrows = True,
          bbox = dict(facecolor = "grey"))

```

redo dag but with the graphviz package

```python
dot = graphviz.Digraph('apriori-dag', comment='aprior dag') 
dot.node('B', 'Biomass', shape = "rect")  
dot.node('C', 'Canopy Closure', shape = "rect")
dot.node('S', 'Stand Age', shape = "rect")
dot.node('A', 'Snow-free date',shape = "rect")
dot.node('R', 'Snow-free duration',  shape = "rect")
dot.node('V', 'Burn Severity', shape = "ellipse", color="red")
dot.node('T', 'Topography',  shape = "rect")
dot.node('D', 'Drought',  shape = "rect")

dot.edges(['BA', 'CA', 'SA', 'AR', 'BV', 'CV', 'SV', 'RV', 'DV', 'TV', 'TA', 'DR'])




dot
```

```python

dot.render(directory='DAGs', view=True)  
```
